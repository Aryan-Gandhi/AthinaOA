# Importing all the dependencies

```python
import os
from langchain_community.document_loaders import PyPDFLoader
from ragas.testset.generator import TestsetGenerator
from ragas.testset.evolutions import simple, reasoning, multi_context
from langchain_openai import ChatOpenAI, OpenAIEmbeddings
import pandas as pd

path = "athina_doc.pdf"

loader = PyPDFLoader(path)
documents = loader.load()
```

## Generating TestSet using RAGA

```python
generator_llm = ChatOpenAI(model="gpt-4")
critic_llm = ChatOpenAI(model="gpt-4")
embeddings = OpenAIEmbeddings()

# Creating an instance of the TestsetGenerator using the specified LLMs and embeddings
generator = TestsetGenerator.from_langchain(
    generator_llm,
    critic_llm,
    embeddings
)

# Defining the distribution of test case types
distributions = {
    simple: 0.5,
    multi_context: 0.4,
    reasoning: 0.1
}

# Generating a test set with 25 test cases based on the given documents and distribution
testset = generator.generate_with_langchain_docs(documents, 25, distributions)
testset.to_pandas()
```
